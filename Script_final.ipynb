```python
import pandas as pd
import numpy as np
```

```python
df_final = pd.read_csv('final.csv', usecols= ['3A', '3H', '3V', '4A', '4H', '4V', 'Condição'])
df_final.head()
```

```python
# Normalização dos dados com sklearn
from sklearn.preprocessing import StandardScaler

# Separando as features
y = df_final['Condição'].values
x = df_final.drop(columns=['Condição']).values
# Normalizando
x = StandardScaler().fit_transform(x)
```

```python
type(x)
```

```python
normalizado = pd.DataFrame(data = x)
df_normalizado = pd.concat([normalizado, df_final[['Condição']].reset_index(drop=True)], axis = 1)
df_normalizado
```

```python
df_normalizado.info()
```

```python
df_normalizado.isnull()
```

```python
df_normalizado = df_normalizado.dropna()
```

```python
df_normalizado['Condição'] = df_normalizado['Condição'].replace('Desbalanceado','1')
df_normalizado['Condição'] = df_normalizado['Condição'].replace('Desalinhamento_Horizontal','2')
df_normalizado['Condição'] = df_normalizado['Condição'].replace('Desalinhamento_Vertical','3')
df_normalizado['Condição'] = df_normalizado['Condição'].replace('Normal','4')
```

```python
df_normalizado.head()
```

```python
df_normalizado.tail()
```

```python
df_normalizado.info()
```

```python
df_normalizado['Condição'] = df_normalizado['Condição'].astype(int)
df_normalizado.info()
```

```python
df_normalizado.info()
```

```python
df_normalizado.tail()
```

```python
df_normalizado.columns = ['3A','3H','3V','4A','4H','4V','Condição']
df_normalizado.head()
```

```python
dados = df_normalizado.drop('Condição', axis='columns')
y = df_normalizado['Condição']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dados, y, test_size=0.3, shuffle=True)
```

```python
print("Shape of Train Data : {}".format(X_train.shape))
print("Shape of Test Data : {}".format(X_test.shape))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3) #n_neighbors = k
knn.fit(X_train,y_train)
print("k={}NN Accuracy on Train Data: {}".format(3,knn.score(X_train,y_train)))
print("k={}NN Accuracy on Test Data: {}".format(3,knn.score(X_test,y_test)))
```
